```python
# Importing Libraries

import pandas as pd
import numpy as np
```

#### Open the ` Employee.csv` file and name your dataset `employee`.

```python
# Importing dataset

employee = pd.read_csv('./data/Employee.csv')
display(employee, employee.shape)
```

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1: Group By

employee.groupby(['Department']).agg({'Salary':np.mean})
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html

#pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)

# table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'], aggfunc={'D': np.mean, 'E': np.mean})
```

```python
# Method 2: Pivot Table

table = pd.pivot_table(employee, values=['Salary'], index=['Department', 'Title'], aggfunc={'Salary': np.mean})#.reset_index()

#Renaming column

table.columns = ['AVG Salary']

table
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
table2 = pd.pivot_table(employee, values=['Years','Salary'], index=['Title'], aggfunc={'Years': np.mean, 'Salary': np.mean}, sort=True)

#Renaming columns
table2.columns = ['AVG Salary', 'AVG Years']

table2
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# Count in aggregate, i.e: df.groupby(["id", "pushid"]).agg({"sess_length": ["sum", "mean", "count"]})
```

```python
table3 = pd.pivot_table(employee, values=['Name','Salary'], index=['Department','Title'], aggfunc={'Name': ["count"], 'Salary': np.mean}, sort=True)

#Renaming columns
table3.columns = ['Num_Employees', 'AVG Salary']

table3
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# We can also calculate multiple types of aggregations for any given value column.

# table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'], aggfunc={'D': np.mean,'E': [min, max, np.mean]})
```

```python
table4 = pd.pivot_table(employee, values=['Salary'], index=['Department','Title'], aggfunc={'Salary': [np.median, min, max, np.std] }, sort=True)

table4
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# We can also fill missing values using the fill_value parameter.

# table = pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'], aggfunc=np.sum, fill_value=0)
```

```python
# Changing Nan values to 0

table5 = pd.pivot_table(employee, values=['Salary'], index=['Department','Title'], aggfunc={'Salary': [np.median, min, max, np.std] }, fill_value=0, sort=True)

table5
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here

table6 = pd.pivot_table(employee, values=['Salary'], index=['Department'], aggfunc={'Salary': sum}, sort=True)

table6.columns = ['Total_Salary']

table6
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
table7 = pd.pivot_table(employee, values=['Years','Salary'], index=['Department','Title'], aggfunc={'Years': sum , 'Salary': np.mean}, sort=True)

# Renaming columns
table7.columns = ['AVG Salary','Total Years']

table7
```
